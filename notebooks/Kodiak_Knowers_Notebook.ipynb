### **2.2 Primary Dataset: smmh.csv**

| Item | Description |
|---|---|
| Source | Kaggle: *Social Media and Mental Health* |
| File used | `smmh.csv` |
| Observations | 481 survey respondents |
| Variables | 21 columns |
| Unit of observation | Each row represents one respondent's response |
| General topic | Social media habits and *self-reported* mental health, sleep, and attention-related experiences |

The dataset contains survey responses about the respondents' demographic, amount of time spent on social media, platform preferences, and self-reported behaviors. Specifically including, using social media without a purpose, distraction, restlessness, worriness, concentration, comparison with others, validation-seeking, feeling depressed, fluctuation on everyday activities, and sleep issues.

The dataset was collected through a survey for a university statistics course project, which focused on different aspects of mental health and how a participant feels any effect on social media use. However, the Kaggle description of the dataset did not specify whether the respondents were randomly selected, how the survey was given, or if the survey was done online or offline. This makes the sampling method unclear and may not fully represent a broad population of young adults. Therefore, the findings from this dataset should be interpreted within the surveyed respondents from the dataset and not towards the population of all young adults.

Since the sampling method was not specified, there is a possibility of having a sampling bias. The dataset may be affected by voluntary response bias or convenience bias. However, it is hard to conclude if there was a bias in this dataset and should be treated as possible limitations because the source of the dataset does not provide enough information about the sampling processes of this dataset.

### **2.3 Attributes of the Primary Dataset**

| Attribute | Columns/Variables | Description |
|---|---|---|
| Response Metadata | Timestamp | Date and time when the survery response was submitted. |
| Demographics | Age, gender, relationship status, occupation status, organization affiliation | Describes the respondent's background and identifies the target group of young adults |
| Social Media access and platforms | Social media use, commonly used platforms | Identifies whether the respondent uses social media and which platforms they use. |
| Social media usage intensity | Average daily time spent on social media | Describes the main variable for comparing heavier-usage and lighter-usage users. |
| Social media behavior patterns | Purposeless use, distracted while bust, restlessness after not using social media | Measures the possible compulsive or disruptive social media behaviors. |
| Productivity-related indicators | Ease of distraction, difficulty concentrating | Used as indicators of productivity or focus difficulties. |
| Mental health indicators | Worry, comparison with others, feelings about comparison, validation-seeking, feeling depressed, fluctuation of interest in daily activities | Used to explore mental health-related patterns associated with social media use. |
| Sleep indicator | Frequency of sleep issues | Used to check the relationship between social media use and sleep-related problems. |
