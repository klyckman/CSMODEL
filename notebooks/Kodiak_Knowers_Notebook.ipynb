# **1.0 Abstract**

This study examines the question “How are social media use patterns associated with mental health, sleep quality, and focus among young adults?”. To explore this, three datasets were obtained and harmonised: smmh.csv (481 respondents, 21 variables) from Kaggle’s Social Media and Mental Health dataset; Mental_Health_and_Social_Media_Balance_Dataset.csv (500 respondents, 10 variables) from Kaggle’s Mental Health & Social Media Balance Dataset; and Social_Media_Impact_on_Daily_Lives.csv (522 respondents, 16 variables) from Mendeley’s The Impact of Social Media on Daily Lives. These datasets contain demographic information, social‑media usage patterns, subjective mental‑health indicators, sleep‑quality measures, stress and happiness scores, and productivity-related behaviours. The datasets involve different respondents and lack a common identifier, so we did not merge them row‑by‑row. The selected variables (age, gender, daily digital media hours, sleep problem score and mental‑health difficulty score) were cleaned and harmonised into a combined dataset. Additionally, all three sources rely on subjective surveys, and their sampling methods are not clearly specified. Therefore, any insights should be interpreted as associations within the surveyed respondents rather than conclusions.

##### **3.2.7.1 Sleep Problem Score**



The sleep_problem_score was combined/harmonized from sleep-related variables in two datasets. In smmh.csv, the original sleep issue scale was already measured sleep difficulty, where higher values meant more sleep problems. In the Mental_Health_and_Social_Media_Balance_Dataset.csv, the original sleep quality variable had to be reversed because higher sleep quality meant better sleep. Both variables were normalized to a 0 to 1 scale so that higher values consistenly represent greater sleep problems. The Social Media Impact on Daily Lives.csv did not contain any variable that related to sleep, so the values in this specific feature would be missing (NaN).

```python
# Harmonize sleep-related variables into one.
# sleep_problem_score uses a 0 to 1 scale:
# 0 = fewer sleep problems / better sleep
# 1 = more sleep problems / worse sleep

df1_combined["sleep_problem_score"] = (pd.to_numeric(df1_clean["sleep_issue_scale"], errors="coerce") - 1) / 4

df2_combined["sleep_problem_score"] = (10 - pd.to_numeric(df2_clean["sleep_quality"], errors="coerce")) / 9

df3_combined["sleep_problem_score"] = np.nan
```

##### **3.2.7.2 Mental Health Difficulty Score**

This is a feature that was created to combine variables that are mental health-related indicators from all datasets. Since the datasets are using different scale ranges, the values would be converted into a 0 to 1 scale, where higher values represent a good mental health-related difficulty.

From df1, the score was made from using the worry_scale, depression_scale, and interest_flucturation variables. These variables were selected because it directly describes subjective mental health-related experiences. From df2, stress level was used as it is, while the happiness index had to be reversed because the higher the scale the happier the happiness is. Then, the reversed happiness score and stress levels were averaged.

Finally, df3 did not contain any variables that was about any mental health scale, so the values from this dataset will be treated as missing (NaN).

```python
# For df1, combine mental health-related variables into one.
# mental_health_difficulty_score uses a 0 to 1 scale:
# 0 = lower mental health difficulty
# 1 = higher mental health difficulty

df1_mental_health_cols = [
    "worry_scale",
    "depression_scale",
    "interest_fluctuation"
]

df1_mental_health_normalized = (df1_clean[df1_mental_health_cols].apply(pd.to_numeric, errors="coerce").sub(1).div(4))

df1_combined["mental_health_difficulty_score"] = df1_mental_health_normalized.mean(axis=1)
```

```python
# For df2, stress score is already the score for stress.
# Happiness score must be reversed because higher happiness means better well-being.

df2_stress_score = (pd.to_numeric(df2_clean["stress_level"], errors="coerce") - 1) / 9

df2_reversed_happiness_score = (10 - pd.to_numeric(df2_clean["happiness_index"], errors="coerce")) / 9

df2_combined["mental_health_difficulty_score"] = pd.concat([df2_stress_score, df2_reversed_happiness_score],axis=1).mean(axis=1)
```

```python
# For df3, it does not contain a health-related score, so this feature is marked as missing for df3 rows.

df3_combined["mental_health_difficulty_score"] = np.nan
```

##### **3.2.7.3 Focus Score**



```python
# Harmonize focus-related variables into one comparable feature.
# focus_difficulty_score uses a 0 to 1 scale:
# 0 = lower focus/time-loss difficulty
# 1 = higher focus/time-loss difficulty

df1_focus_cols = [
    "distraction_frequency",
    "distraction_scale",
    "concentration_difficulty"
]

df1_focus_normalized = (
    df1_clean[df1_focus_cols]
    .apply(pd.to_numeric, errors="coerce")
    .sub(1)
    .div(4)
)

df1_combined["focus_difficulty_score"] = df1_focus_normalized.mean(axis=1)
```

```python
# For df2, it does not contain a focus-related score, so this feature is marked as missing for df2 rows.
df2_combined["focus_difficulty_score"] = np.nan
```

**Descriptive Statistics** - This table summarizes the central tendency and spread of both variables. There is also a somewhat big standard deviation, which indicates a wide variation in media use. ### **4.1.2 Is higher digital/social media use associated with poorer sleep outcomes?**

This analysis uses the harmonized `sleep_problem_score`, where higher values indicate worse sleep outcomes. Rows with missing sleep scores were excluded only for this EDA question because those missing values come from datasets that did not collect sleep-related variables.

```python
# Create a sleep-specific EDA dataframe using only rows with sleep information.
# This avoids imputing values for datasets that did not collect sleep-related variables.

sleep_eda = combined_dfs.dropna(
    subset=["daily_digital_media_hours", "sleep_problem_score"]
).copy()

# Create usage groups for clearer comparison.
def classify_usage(hours):
    if pd.isna(hours):
        return np.nan
    elif hours < 2:
        return "Low use"
    elif hours < 4:
        return "Moderate use"
    else:
        return "High use"

sleep_eda["usage_group"] = sleep_eda["daily_digital_media_hours"].apply(classify_usage)

# Check how many rows remain per source.
sleep_eda["source_dataset"].value_counts()
```

Rows with missing sleep scores were removed only for this EDA because those rows came from datasets that did not collect sleep-related variables. These values were not imputed because doing so would create artificial sleep scores.

#### **Numerical Summaries**

```python
sleep_summary = sleep_eda.groupby("usage_group")["sleep_problem_score"].agg(["count", "mean", "median", "std"]).reindex(["Low use", "Moderate use", "High use"]).round(3)

sleep_summary
```

#### **Spearman Correlation**

```python
sleep_spearman = sleep_eda["daily_digital_media_hours"].corr(
    sleep_eda["sleep_problem_score"],
    method="spearman"
)

print(f"Spearman correlation between digital/social media hours and sleep problem score: {sleep_spearman:.3f}")
```

#### **Data Visualization**

```python
plt.figure(figsize=(8, 5))

sns.boxplot(
    data=sleep_eda,
    x="usage_group",
    y="sleep_problem_score",
    order=["Low use", "Moderate use", "High use"]
)

sns.stripplot(
    data=sleep_eda,
    x="usage_group",
    y="sleep_problem_score",
    order=["Low use", "Moderate use", "High use"],
    color="red",
    alpha=0.25,
    jitter=True
)

plt.title("Sleep Problems by Digital/Social Media Use Level")
plt.xlabel("Digital/Social Media Use Level")
plt.ylabel("Sleep Problem Score")
plt.tight_layout()
plt.show()
```

The boxplot compares normalized sleep problem scores across low, moderate, and high digital/social media use groups. The three groups have very similar median sleep problem scores of 0.50, and their mean scores are also close to each other. The Spearman correlation between daily digital/social media hours and sleep problem score is 0.038, which suggests a very weak or nearly nonexistent association. Based on this EDA, heavier digital/social media use does not appear to be strongly associated with poorer sleep outcomes in the harmonized sample.

```python
### Quick cell of code to shorten the dataset source names

sleep_eda["source_label"] = sleep_eda["source_dataset"].replace({
    "smmh.csv": "SMMH",
    "Mental_Health_and_Social_Media_Balance_Dataset.csv": "MH-Social Balance"
})
```

However, when the data is separated by source datasets, the sleep problem scores appear to differ more by dataset source than by usage group. The `smmh.csv` dataset responses show higher and more widely dispersed sleep problem scores, while the `Mental Health and Social Media Balance.csv` dataset generally shows lower scores. This suggests that differences in survey design and measurement may influence the harmonized sleep variable. ### **4.1.3 Is higher digital/social media use associated with greater difficulty in maintaining focus and concentration?**

This section examines whether higher digital/social media use is associated with greater difficulty in maintaining focus and concentration. The analysis uses `daily_digital_media_hours` as the measure of social media use and `focus_difficulty_score` as the outcome variable. The data is filtered to include only rows with valid values for both variables, then grouped into low, moderate, and high usage categories to compare focus difficulty across different levels of use. Pearson and Spearman correlations are also computed to check whether the relationship is linear or monotonic. This section helps determine whether heavier digital/social media use is linked to worse focus outcomes.

```python
# Prepare focus EDA dataset (using all available focus data)
focus_eda = combined_dfs[['daily_digital_media_hours', 'focus_difficulty_score']].dropna()

print(f"Sample size for focus analysis: {len(focus_eda)}")
print("\nDescriptive Statistics:")
print(focus_eda.describe())
```

```python
# Create usage groups
focus_eda['usage_group'] = pd.cut(focus_eda['daily_digital_media_hours'],
                                   bins=[0, 3, 5, 11],
                                   labels=['Low use', 'Moderate use', 'High use'])

# Summary statistics by usage group
focus_summary = focus_eda.groupby('usage_group')['focus_difficulty_score'].agg(['count', 'mean', 'median', 'std'])

print("Focus Difficulty Score by Usage Group:")
print(focus_summary)
```

```python
# Pearson correlation
focus_pearson = focus_eda['daily_digital_media_hours'].corr(
    focus_eda['focus_difficulty_score'],
    method='pearson'
)

print(f"Pearson correlation between digital/social media hours and focus difficulty score: {focus_pearson:.3f}")
```

```python
# Spearman correlation
focus_spearman = focus_eda['daily_digital_media_hours'].corr(
    focus_eda['focus_difficulty_score'],
    method='spearman'
)

print(f"Spearman correlation between digital/social media hours and focus difficulty score: {focus_spearman:.3f}")
```

#### **4.1.3.1 Data Visualization**

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
plt.scatter(focus_eda['daily_digital_media_hours'], focus_eda['focus_difficulty_score'], alpha=0.5)
z = np.polyfit(focus_eda['daily_digital_media_hours'], focus_eda['focus_difficulty_score'], 1)
p = np.poly1d(z)
plt.plot(focus_eda['daily_digital_media_hours'].sort_values(), p(focus_eda['daily_digital_media_hours'].sort_values()), "r--", alpha=0.8, label=f'Trend (r={focus_pearson:.3f})')
plt.xlabel('Daily Digital Media Hours')
plt.ylabel('Focus Difficulty Score (0-1)')
plt.title('Relationship between Digital Media Use and Focus Difficulty')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Box plot by usage group
plt.figure(figsize=(10, 6))
focus_eda.boxplot(column='focus_difficulty_score', by='usage_group', figsize=(10, 6))
plt.title('Focus Difficulty Score by Social Media Usage Group')
plt.suptitle('') 
plt.xlabel('Usage Group')
plt.ylabel('Focus Difficulty Score (0-1)')
plt.tight_layout()
plt.show()
```

```python
# Distribution plots for each usage group
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, group in enumerate(['Low use', 'Moderate use', 'High use']):
    data = focus_eda[focus_eda['usage_group'] == group]['focus_difficulty_score']
    axes[idx].hist(data, bins=15, edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'{group}\n(n={len(data)})')
    axes[idx].set_xlabel('Focus Difficulty Score')
    axes[idx].set_ylabel('Frequency')
    axes[idx].axvline(data.mean(), color='red', linestyle='--', label=f'Mean: {data.mean():.3f}')
    axes[idx].legend()

plt.tight_layout()
plt.show()
```

The results suggest a weak-to-moderate positive relationship between digital/social media use and focus difficulty. The sample size was 817, and both correlations were positive: `Pearson = 0.222` and `Spearman = 0.270`. The usage-group summary also shows a rise in focus difficulty from `low use (mean = 0.514)` to `moderate use (0.557)` and `high use (0.688)`, which supports the idea that heavier users tend to report more difficulty concentrating. However, the relationship is still not strong, so social media use explains only part of the variation in focus difficulty.

## **Summary, Conclusions, and Recommendations**

**Summary**

This study examined how social media use is related to mental health, sleep quality, and focus among young adults aged 18-30. Three different survey datasets were cleaned, standardized and combined into one dataset with 988 respondents. Since the datasets came from different people and had no shared identifiers, they were not merged directly. Instead, important variables such as daily digital media use, mental health difficulty, sleep problems and focus difficulty were converted into comparable scales.

Exploratory data analysis (EDA) was conducted using descriptive statistics, visualizations, and correlation tests (Pearson and Spearman). The results were used to examine whether higher digital media use is associated with worse mental health, sleep, and focus outcomes.

The dataset established a mental health baseline across the three combined files to analyze general psychological well-being alongside screentime.

The study found almost no link between daily screen hours and sleep issues. The Spearman correlation value was extremely low at 0.038. Looking at the groups, the average sleep problem scores were nearly identical across all users.

- Low users: 0.536
- Moderate users: 0.522
- High users: 0.545.

For Focus and Concentration Difficulty, there is a weak-to-moderate positive relationship between media use and concentration problems. Across a sample size of 817 respondents, both statistical correlations were positve (Pearson = 0.222 and Spearman = 0.270). The data clearly shows that focus difficulties like distraction increase as screen time goes up.

- Low use average: 0.514
- Moderate use average: 0.557,
- High use average: 0.688.


**Conclusions**

1. Social media use has a very weak relationship with mental health difficulties.

The correlation result (ρ = 0.078) shows that increased social media use is only slightly linked with higher mental health difficulty. However, the relationship is very small, meaning social media use alone does not strongly explain mental health differences among respondents.

2. Social media use is almost not related to sleep problems.

The correlation (ρ = 0.038) is extremely low, which suggests that higher or lower social media use does not clearly affect sleep quality in this dataset. Sleep results also looked similar across low, moderate, and high usage groups.

3. Social media use has a weak to moderate relationship with focus problems.

The strongest finding in this study is for focus and concentration. Both Pearson (r = 0.222) and Spearman (ρ = 0.270) show a positive relationship, meaning higher social media use is associated with more difficulty focusing. Although not strong, this trend is more consistent compared to the other variables.




Overall, EDA suggests that social media does not directly influence mental health, sleep problems, or focus problems directly. Since all correlations are small to moderate, it suggests that other factors (like lifestyle, stress, study/work load, and personal habits) likely have a bigger role in mental health, sleep, and focus outcomes.

Because the data is based on self-reported surveys, we cannot say that social media use directly causes changes in mental health, sleep, or focus. The results only show relationships between variables.


**Recommendations**

1. Researchers should ensure that variables are measured in a consistent way across studies. In particular, digital media use should be clearly defined (e.g., social media time vs total screen time), and mental health and sleep indicators should use comparable scales. This will improve the reliability of combined analyses.

2. Future studies may include additional regression analysis such as: academic performance, lifestyle habits (diet, physical activity), stress triggers outside social media to help better explain mental health and productivity outcomes.
