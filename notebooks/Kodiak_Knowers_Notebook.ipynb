# **2. Dataset Description**

## **2.1 Dataset Selection**

The study will be using a total of three datasets, *smmh.csv*, *Mental_Health_and_Social_Media_Balance_Dataset.csv*, and *Social Media Impact on Daily Lives.csv*, which will be merged to obtain insightful information about social media use, mental health, sleep quality, and productivity-related behaviors. The *smmh.csv* and *Mental_Health_and_Social_Media_Balance_Dataset.csv* datasets were obtained from the Kaggle datasets **Social Media and Mental Health** and **Mental Health & Social Media Balance Dataset**, respectively. Whereas the *Social Media Impact on Daily Lives.csv* was obtained from the Mendelay dataset **The Impact of Social Media on Daily Lives**.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df1_raw = pd.read_csv("../data/raw/smmh.csv")
df2_raw = pd.read_csv("../data/raw/Mental_Health_and_Social_Media_Balance_Dataset.csv")
df3_raw = pd.read_csv("../data/raw/Social_Media_Impact_on_Daily_Lives.csv")
```

# DF1 cleaning, renaming, and dropping

```python
df1_clean = df1_raw.copy()

df1_clean = df1_clean.drop(columns = ['Timestamp',
                      '5. What type of organizations are you affiliated with?',
                      '3. Relationship Status'])

df1_clean = df1_clean.rename(columns={
    '1. What is your age?': 'age',
    '2. Gender': 'gender',
    '4. Occupation Status': 'occupation_status',
    '6. Do you use social media?': 'uses_social_media',
    '7. What social media platforms do you commonly use?': 'platforms_used',
    '8. What is the average time you spend on social media every day?': 'daily_time_spent',
    '9. How often do you find yourself using Social media without a specific purpose?': 'purposeless_usage',
    '10. How often do you get distracted by Social media when you are busy doing something?': 'distraction_frequency',
    '11. Do you feel restless if you haven\'t used Social media in a while?': 'restlessness_score',
    '12. On a scale of 1 to 5, how easily distracted are you?': 'distraction_scale',
    '13. On a scale of 1 to 5, how much are you bothered by worries?': 'worry_scale',
    '14. Do you find it difficult to concentrate on things?': 'concentration_difficulty',
    '15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?': 'social_comparison_frequency',
    '16. Following the previous question, how do you feel about these comparisons, generally speaking?': 'comparison_sentiment',
    '17. How often do you look to seek validation from features of social media?': 'validation_seeking',
    '18. How often do you feel depressed or down?': 'depression_scale',
    '19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?': 'interest_fluctuation',
    '20. On a scale of 1 to 5, how often do you face issues regarding sleep?': 'sleep_issue_scale'
})

df1_clean
```

```python
# Remove rows where respondent answered "no" in 'uses_social_media' column.
df1_clean = df1_clean[df1_clean["uses_social_media"].str.strip().str.lower() == "yes"].copy()

df1_clean["uses_social_media"].value_counts(dropna=False)
```

```python
# Drop column since we now filtered out all non-social media users
df1_clean = df1_clean.drop(columns=["uses_social_media"])

df1_clean
```

```python
# Filter age range of DF1 (18-30)
df1_clean["age"] = pd.to_numeric(df1_clean["age"], errors="coerce")

df1_clean = df1_clean[(df1_clean["age"] >= 18) & (df1_clean["age"] <= 30)].copy()

df1_clean
```

# DF2 cleaning, renaming, and dropping

```python
df2_clean = df2_raw.copy()

df2_clean = df2_clean.drop(columns=["User_ID"])

df2_clean = df2_clean.rename(columns={
    "Age": "age",
    "Gender": "gender",
    "Daily_Screen_Time(hrs)": "daily_screen_time_hours",
    "Sleep_Quality(1-10)": "sleep_quality",
    "Stress_Level(1-10)": "stress_level",
    "Days_Without_Social_Media": "days_without_social_media",
    "Exercise_Frequency(week)": "exercise_frequency_week",
    "Social_Media_Platform": "primary_social_media_platform",
    "Happiness_Index(1-10)": "happiness_index"
})

df2_clean
```

```python
# Set age range
df2_clean["age"] = pd.to_numeric(df2_clean["age"], errors="coerce")

df2_clean = df2_clean[(df2_clean["age"] >= 18) & (df2_clean["age"] <= 30)].copy()

df2_clean
```

# DF3 cleaning, dropping, and renaming

```python
df3_clean = df3_raw.copy()

df3_clean = df3_clean.drop(columns=["Unnamed: 0",
                                    "Enter your name",
                                    "Do you think privacy policies are effective in social networking sites?",
                                    "Does social media advertising influence the purchase of your product?"])

df3_clean = df3_clean.rename(columns={
    'Enter your age (Ex: 20)': 'age',
    'What is your gender': 'gender',
    'What is your profession?': 'occupation',
    'On an average how many hours a day you spend on your profession?': 'daily_occupation_hours',
    'What social networking sites do you use?': 'social_media_platforms',
    'On an average how many hours a day do you spend on these sites?': 'daily_social_media_hours',
    'On an average how many hours a day do you spend with your family?': 'daily_family_hours',
    'Why do you use these social networking sites?': 'social_media_use_reasons',
    'What are personal benefits of  using social networking sites?': 'perceived_social_media_benefits',
    'When do you access social media websites?': 'times_using_social_media',
    'When you use social media did you forget about time?': 'forget_time_using_social_media',
    'Do you consider yourself addicted to social media?': 'subjective_social_media_addiction'
})

df3_clean
```

```python
# Set age range
df3_clean["age"] = pd.to_numeric(df3_clean["age"], errors="coerce")

df3_clean = df3_clean[(df3_clean["age"] >= 18) & (df3_clean["age"] <= 30)].copy()

df3_clean
```

# Combining DFs

```python
df1_clean["source_dataset"] = "smmh.csv"
df2_clean["source_dataset"] = "Mental_Health_and_Social_Media_Balance_Dataset.csv"
df3_clean["source_dataset"] = "Social_Media_Impact_on_Daily_Lives.csv"
```

```python
# DF1: social media time ranges
df1_time_mapping = {
    "Less than an Hour": 0.5,
    "Between 1 and 2 hours": 1.5,
    "Between 2 and 3 hours": 2.5,
    "Between 3 and 4 hours": 3.5,
    "Between 4 and 5 hours": 4.5,
    "More than 5 hours": 5.5
}

# DF3: social media time ranges
df3_time_mapping = {
    "1-2": 1.5,
    "3-4": 3.5,
    "4-5": 4.5,
    "5-6": 5.5,
    "6+": 6
}
```

```python
df1_combined = pd.DataFrame({
    "source_dataset": "smmh.csv",
    "age": df1_clean["age"],
    "gender": df1_clean["gender"].str.strip().str.lower(),
    "daily_digital_media_hours": df1_clean["daily_time_spent"].map(df1_time_mapping),
    "time_measure_type": "social_media_time",
    "platform_info": df1_clean["platforms_used"],
    "platform_count": df1_clean["platforms_used"].str.split(",").str.len(),
    "platform_info_type": "multiple_platforms"
})

df2_combined = pd.DataFrame({
    "source_dataset": "Mental_Health_and_Social_Media_Balance_Dataset.csv",
    "age": df2_clean["age"],
    "gender": df2_clean["gender"].str.strip().str.lower(),
    "daily_digital_media_hours": df2_clean["daily_screen_time_hours"],
    "time_measure_type": "screen_time",
    "platform_info": df2_clean["primary_social_media_platform"],
    "platform_count": 1,
    "platform_info_type": "primary_platform_only"
})

df3_combined = pd.DataFrame({
    "source_dataset": "Social_Media_Impact_on_Daily_Lives.csv",
    "age": df3_clean["age"],
    "gender": df3_clean["gender"].str.strip().str.lower(),
    "daily_digital_media_hours": df3_clean["daily_social_media_hours"].map(df3_time_mapping),
    "time_measure_type": "social_media_time",
    "platform_info": df3_clean["social_media_platforms"],
    "platform_count": df3_clean["social_media_platforms"].str.split(",").str.len(),
    "platform_info_type": "multiple_platforms"
})
```

```python
combined_dfs = pd.concat(
    [df1_combined, df2_combined, df3_combined],
    ignore_index=True
)

combined_dfs.head()
```

## 4.0 **Research Question & Exploratory Data Analysis**

### 4.1 **EDA Questions**

To address our research question regarding the associations between social media patterns, mental health, and productivity, we will explore:

1.) What are the common daily screen times and social media platforms used by young adults across different datasets?

2.) Is there an association between higher social media usage (hours) and reported sleep quality?

3.) Is there a relationship between the time spent on social media and perceived social media addiction?

### 4.1.1 **What are the common daily screen times and social media platforms used by young adults across different datasets?**

To start, a histogram was implemented that reveals the distribution shape of daily screen time is positively skewed (right-skewed), indicating a central tendency where the mode and median fall between 3 and 5 hours, but a long tail of extreme users pulls the mean higher. The bar plot confirms Facebook and YouTube as the most frequent modes of platform usage. The overall dispersion shows significant variance, confirming a broad spectrum of digital habits among the surveyed young adults.

```python
sns.set_theme(style="whitegrid")

# 2 subplots: one for daily screen time distribution, one for top social media platforms
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1
sns.histplot(data=combined_dfs, x="daily_digital_media_hours", bins=10, kde=True, color="teal", ax=axes[0])
axes[0].set_title("Distribution of Daily Screen Time")
axes[0].set_xlabel("Hours per Day")
axes[0].set_ylabel("Count of Respondents")

# Plot 2
platforms = combined_dfs['platform_info'].dropna().str.split(',').explode().str.strip()

# Standardize capitalization
platforms = platforms.str.title()
top_platforms = platforms.value_counts().head(8)

sns.barplot(x=top_platforms.values, y=top_platforms.index, palette="viridis", ax=axes[1])
axes[1].set_title("Top Social Media Platforms")
axes[1].set_xlabel("Number of Users")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

### 4.1.2 **Is there an association between higher social media usage (hours) and reported sleep quality?**

To address Question 2, we utilized a heatmap to calculate the Pearson correlation across multiple continuous variables, specifically targeting the relationship between screen time and sleep quality. The matrix reveals a negative correlation between the two. This indicates that as the dispersion of screen time shifts toward higher hours, the central tendency of reported sleep quality decreases.

```python
plt.figure(figsize=(8, 6))

# Select the appropriate columns for correlation analysis
cols_to_correlate = ['daily_screen_time_hours', 'sleep_quality', 'stress_level', 'happiness_index']

# Calculate the Pearson correlation
corr_matrix = df2_clean[cols_to_correlate].corr(method='pearson')

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)

plt.title("Pearson Correlation Heatmap:\nScreen Time vs. Sleep & Mental Health")
plt.tight_layout()
plt.show()
```

### 4.1.3 **Is there a relationship between the time spent on social media and perceived social media addiction?**

For the last EDA question, we utilized a boxplot and calculated the Spearman correlation to evaluate the monotonic relationship between daily screen time and the ordinal addiction tiers. The positive correlation reveals that as daily social media hours increase, the central tendency of perceived addiction also rises. The dispersion of hours is noticeably wider among respondents answering "sometimes" or "yes", whereas users claiming "no" addiction tightly cluster around lower usage intervals.

```python
plt.figure(figsize=(9, 5))

# Map the text answers to numbers
addiction_mapping = {'no': 0, 'not sure': 1, 'sometimes': 2, 'yes': 3}
df3_clean['addiction_score'] = df3_clean['subjective_social_media_addiction'].str.lower().map(addiction_mapping)

# Convert into actual numbers so we can perform correlation analysis
def convert_hours_to_num(val):
    if pd.isna(val): return np.nan
    val = str(val).strip()
    if val == '1-2': return 1.5
    elif val in ['2-4', '3-4']: return 3.0
    elif val == '4-5': return 4.5
    elif val == '4-6': return 5.0
    elif val == '6-8': return 7.0
    elif val == '6+': return 6.5
    else: return np.nan

# Creates the _num column
df3_clean['daily_social_media_hours_num'] = df3_clean['daily_social_media_hours'].apply(convert_hours_to_num)

# Calculate the Spearman correlation
spearman_corr = df3_clean['daily_social_media_hours_num'].corr(df3_clean['addiction_score'], method='spearman')

# Boxplot for the distribution
sns.boxplot(data=df3_clean, 
            x="subjective_social_media_addiction", 
            y="daily_social_media_hours_num", 
            order=['no', 'not sure', 'sometimes', 'yes'], 
            palette="Set2")

plt.title(f"Distribution: Daily Social Media Hours vs. Perceived Addiction\n(Spearman Correlation: {spearman_corr:.2f})")
plt.xlabel("Self-Perceived Addiction Level")
plt.ylabel("Daily Social Media Time (Hours)")

plt.tight_layout()
plt.show()
```
